# Graphing in `pandas` with `matplotlib`

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
small_df = pd.read_csv('13-classwork/countries.csv')
```

```python
gender_df = pd.read_csv('13-classwork/gender.csv')
```

```python
ax = gender_df.plot(kind='barh', x='state', stacked=True, figsize=(10,3))
ax.set_title('Test, test', pad=40)
ax.title.set_fontsize(30)
ax.title.set_fontname('Source Sans Pro')
```

```python

color_dict = {
  'Antigua and Barbuda': 'lightgray',
  'Angola': 'lime',
  'Algeria': 'lightgray',
  'Albania': 'lightgray',
  'Afghanistan': 'lightgray'
}

bar_colors = [color_dict[name] for name in small_df['country'][:5]]

ax = small_df.head(5).plot(x='country', y='life_expectancy', kind='barh', 
                      color=bar_colors, legend=False)
ax.set_xticks([0, 20,40,60,80]) # location, where ticks appear
# ax.set_xlabel('Life expectancy (years)')
ax.set_xticklabels(['Tiny baby', 20, 40, 60, '80 years old']) # what the 
# labels should read

```
