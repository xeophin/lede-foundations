## The three things you should do when you read in a dataset

1. Get an overview of this data using `.head()`
2. Count rows and columns using `.shape`
3. Look at the datatypes using `.dtypes`

```python
df = pd.read_csv('05-classwork/countries.csv')
```

```python
df.head()
```

```python
df.shape
```

```python
df.dtypes
```

## Analysis based on Categories

```python

df.groupby('Continent')
```

```python
df.groupby('Continent').life_expectancy.median()
```

```python
df.groupby('Continent').Population.sum()
```

```python
# sort_values will automatically sort based on the values if we don't have a 
# data frame
df.groupby('Continent').Population.sum().sort_values(ascending=False)
```

```python
df.Continent.value_counts()
```

```python
# By using 'normalize', you can automatically get percentages
df.Continent.value_counts(normalize=True)
```

When working with data frames, no loops are necessary in order to manipulate all values within the frame.

```python
df.Continent.value_counts(normalize=True) * 100 
```

```python
df.GDP_per_capita * df.Population
```

```python
# Data can be stored in new columns
# This requires bracket notation, dot notation will not work
df['GDP'] = df.GDP_per_capita * df.Population
```

```python
df.head(3)
```

## Filter the list

```python
df[df.Continent == 'Africa']
```

```python
df[(df.Continent == 'Africa') & (df.life_expectancy > 60)]
```

## Plotting data

```python
# Magic command for Jupyter Notebook
%matplotlib inline

df.groupby('Continent').life_expectancy.median().plot(kind='bar')
```

```python
df.groupby('Continent').life_expectancy.median().plot(kind='barh')
```

```python
df.groupby('Continent').life_expectancy.median().sort_values(
  ascending=False).plot(
  kind='barh')
```

```python
df.plot(x='GDP_per_capita', y='life_expectancy', kind='scatter')
```
