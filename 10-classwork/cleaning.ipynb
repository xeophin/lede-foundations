```python
yellowpages = open('10-classwork/cleaning/data/yellowpages.txt').read()
```

```python
import re
import pandas as pd
matches = re.findall(r'.*\d\d\d-\d\d\d-\d\d\d\d', yellowpages)
df = pd.DataFrame({'full': matches})
```

```python
df.head()
```

```python
df['phone'] = df['full'].str.extract('(\d{3}-\d{3}-\d{4})')
```

```python
df.head()
```

```python
df['address'] = df['full'].str.extract('(^.*?) \.')
```
