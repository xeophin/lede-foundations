```python
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
df = pd.read_csv('14-classwork/visas-by-continent-long.csv')
df.head()
```

```python
df['year_date'] = pd.to_datetime(df['year'], format='%Y')
```

```python
df.head()
```

```python
df.groupby('continent').plot(x='year_date', y='visas')
```

```python
fig, ax = plt.subplots()

df.groupby('continent').plot(x='year_date', y='visas', ax=ax)
```

```python
# converting long data to wide data
```

```python
df.groupby(['year', 'continent'])['visas'].sum().unstack()
# using unstack() changes the data from long to wide format
```
