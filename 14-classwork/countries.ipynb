```python
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
df = pd.read_csv('14-classwork/countries.csv')
df.head()
```

```python
df.plot(x='gdp_per_capita', y='life_expectancy', kind='scatter')
```

```python
fig, ax = plt.subplots()

df.groupby('continent').plot(x='gdp_per_capita', y='life_expectancy', 
                             kind='scatter', ax=ax)
```

```python
fig, ax = plt.subplots()

df.groupby('continent').plot(x='gdp_per_capita', y='life_expectancy',
                             marker='o', linestyle='', ax=ax)
```

```python
df.plot(x='gdp_per_capita', y='life_expectancy', kind='scatter', s=1)
```

```python
df.plot(x='gdp_per_capita', y='life_expectancy', kind='scatter', 
        s=df['population']*0.0000001)
```

```python
groups = df.groupby('continent')
```

```python
for name, group in groups:
  print(name)
```

```python
# creates an empty graph
fig, ax = plt.subplots()

for name, group in groups:
  group.plot(x='gdp_per_capita', y='life_expectancy', kind='scatter',
             s=group['population'] * 0.0000001, label=name, ax=ax)
```

```python
colors = ['#15F', '#5F1', '#F51', '#', 'Peach', 'Crimson', 'Brick']
```

```python
# creates an empty graph
fig, ax = plt.subplots()

for name, group in groups:
  group.plot(x='gdp_per_capita', y='life_expectancy', kind='scatter',
             s=group['population'] * 0.0000001, label=name, ax=ax, 
             color=colors.pop())
```
