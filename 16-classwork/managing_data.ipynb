# Merging Data

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
cities = pd.read_csv('16-classwork/cities-population.csv')
cities.head()
```

```python
states = pd.read_csv('16-classwork/cities-states.csv')
states.head()
```

```python
# merging data frames, by giving on what column the merge should happen

cities_states = cities.merge(states, left_on='city', right_on='city_name')
```

```python
cities_states
```

```python
# drop columns that aren't necessary
cities_states.drop(columns=['city_name'], inplace=True)
```

```python
cities_states
```

```python
world = pd.read_csv('16-classwork/cities-world.csv')
world
```

```python
# stack two dataframes on top of each other
everywhere = pd.concat([world, cities_states], sort=True, ignore_index=True)
```

```python
everywhere
```

```python
#Â Fill in missing data, using `fillna()`

everywhere['country'].fillna('United States', inplace=True)
```

```python
everywhere
```

```python
# Grab the data from people to fill NA
everywhere['population'].fillna(everywhere['people'], inplace=True)
everywhere
```

```python
everywhere.drop(columns=['people', 'state_abbr'], inplace=True)
everywhere
```

```python
world.plot(x='city', y='people', kind='bar')
```

```python
plt.rcParams.update({
  'axes.formatter.limits': [-20,20]
})

world.plot(x='city', y='people', kind='bar')
```

```python
from matplotlib.ticker import FuncFormatter


def with_words(value, pos):
  millions = int(value / 1000000)
  return '{} millions'.format(millions)


formatter = FuncFormatter(with_words)

ax = world.plot(x='city', y='people', kind='bar')

ax.yaxis.set_major_formatter(formatter)
```

```python
def with_commas(value, pos):
  return '{:,}'.format(int(value))

formatter = FuncFormatter(with_commas)

ax = world.plot(x='city', y='people', kind='bar')

ax.yaxis.set_major_formatter(formatter)
```
