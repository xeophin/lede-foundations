```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('16-classwork/countries.csv')
```

```python
ax = df.groupby('continent').life_expectancy.median().plot(kind='barh')
```

```python
ax = df.groupby('continent').life_expectancy.median().plot(kind='barh', 
                                                           color='red')

# to get the index, use enumerate()
for i, bar in enumerate(ax.patches):
  print(i)
  val = int(round(bar.get_width()))
  x_pos = bar.get_width() - 1
  y_pos = i
  ax.text(x_pos, y_pos, val, va='center', ha='right', color='white', 
          fontweight='bold')
```

```python
ax = df.groupby('continent').life_expectancy.median().plot(kind='bar',
                                                           color='red')

# to get the index, use enumerate()
for i, bar in enumerate(ax.patches):
  print(i)
  val = int(round(bar.get_height()))
  x_pos = bar.get_height() - 8
  y_pos = i
  ax.text(y_pos, x_pos, val, va='bottom', ha='center',
          fontweight='bold', color='white')
```

```python
ax = df.groupby('continent').life_expectancy.median().plot(kind='bar',
                                                           color='red')

ax.set_title('Blupp', pad=40, fontsize=24, fontweight=200, ha='left', x=0)

# to get the index, use enumerate()
for i, bar in enumerate(ax.patches):
  print(i)
  val = int(round(bar.get_height()))
  x_pos = bar.get_height() - 8
  y_pos = i
  ax.text(y_pos, x_pos, val, va='bottom', ha='center',
          fontweight='bold', color='white')
  

# get the figure
fig = plt.gcf()

fig.text(0.12, 0.95, 'This is a subtitle', fontsize=14, ha='left')
```
